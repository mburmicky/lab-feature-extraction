7) Define a function that differentiates between continuous and discrete variables. Hint: The number of unique values might be useful. Store continuous data into a continuous_df variable and do the same for discrete_df

```python
#After analyzing the data, we have decided to keep the variables with more than 200 unique values to be continous and rest
#to be discrete

def segmenting_numerical(df):
    dis_cols = []
    con_cols = []
    for cols in numerical_df.columns:
        if numerical_df[cols].nunique() > 200:
            con_cols.append(cols)
        else:
            dis_cols.append(cols)
    discrete_df = numerical_df[dis_cols]
    continuous_df = numerical_df[con_cols]
    return dis_cols, con_cols
```

```python
segmenting_numerical(numerical_df)
```

```python
dis_cols, con_cols = segmenting_numerical(numerical_df)
```

```python
#Redefining continuous_df since the columns were not stored in the newly created data frames
continuous_df = numerical_df[con_cols]
continuous_df
```

```python
#Redefining discrete_df since the columns were not stored in the newly created data frames
discrete_df = numerical_df[dis_cols]
discrete_df
```

9) Check number of unique values in each column, can they be combined in any way to ease encoding? Comment your thoughts and make those changes.

```python
#The three below can be combined into Luxury
value_mapping = {'Sports Car': 'Luxury','Luxury SUV': 'Luxury','Luxury Car': 'Luxury'}
categorical_df['vehicle_class'] = categorical_df['vehicle_class'].replace(value_mapping)
```

```python
#Medical leave, Disabled, and Retired, can be combined to Other, since all 3 have less than 450 values, while Employeed and Unemployed have > 2.2k values
value_mapping = {'Medical Leave': 'Other','Disabled': 'Other','Retired': 'Other'}
categorical_df['employmentstatus'] = categorical_df['employmentstatus'].replace(value_mapping)
```

```python
#The two below can be combined into Master-Doctor
value_mapping = {'Master':'Master-Doctor','Doctor': 'Master-Doctor'}
categorical_df['education'] = categorical_df['education'].replace(value_mapping)
```
